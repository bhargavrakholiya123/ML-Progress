```python
from google.colab import userdata
MY_API = userdata.get('GEMINI')
```

```python
import google.generativeai as genai
genai.configure(api_key = MY_API)
```

```python
def get_embedding(text):
    result = genai.embed_content(
        model="models/text-embedding-004",
        content=text
    )
    return result['embedding']
```

```python
print(get_embedding("I love playing cricket.")[:5])
```

```python
import numpy as np

def cosine_similarity(a, b):
    a = np.array(a)
    b = np.array(b)
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
e1 = get_embedding("Dogs are wonderful pets.")
e2 = get_embedding("Cats are great companions.")
e3 = get_embedding("The stock market crashed today.")

print("Similarity (dogs vs cats):", cosine_similarity(e1, e2))
print("Similarity (dogs vs stock market):", cosine_similarity(e1, e3))
```
