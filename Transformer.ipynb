```python
from transformers import pipeline

# 2. Load the "Pipeline"
# This automatically downloads a pre-trained BERT-style model
classifier = pipeline("sentiment-analysis")

# 3. Use it!
result = classifier("I was expecting the movie to be boring, but it was actually a masterpiece.")
print(result)
```

```python
# 1. Create a "Text Generation" pipeline
# We explicitly ask for 'gpt2' this time
generator = pipeline('text-generation', model='gpt2')

# 2. Give it a prompt
prompt = "The secret to learning Artificial Intelligence is"

# 3. Let it write!
# max_length=50: Stop after 50 words
# num_return_sequences=1: Just give me one story
result = generator(prompt, max_length=50, num_return_sequences=1)

print(result[0]['generated_text'])
```
