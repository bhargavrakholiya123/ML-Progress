```python
from google.colab import userdata
API = userdata.get('API_KEY')
```

```python
import google.generativeai as genai
genai.configure(api_key="API")
```

```python

model = genai.GenerativeModel("gemini-2.5-flash")

```

```python
def explain_code(code):
  prompt = f""" you are proffesional coder and have 10 years of experience in teaching computer science , your task is to explain code most easiest
  and most simplest way
  - make it short and simple
  - give real world example
  - make coding interesting

  code : {code}

  """

  response = model.generate_content(prompt)
  return response.text
```

```python
print(explain_code("""
def add(a, b):
    return a + b
"""))
```
