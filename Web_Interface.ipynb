```python
# 1. Install Gradio
!pip install gradio

import gradio as gr
from transformers import pipeline

# 2. Load the Model (The same one we just used)
generator = pipeline('text-generation', model='gpt2')

# 3. Define the Function
# This is the logic: Input (Text) -> Output (Story)
def generate_story(prompt):
    result = generator(prompt, max_length=100, num_return_sequences=1)
    return result[0]['generated_text']

# 4. Create the Interface
# We tell Gradio: "Create a box for Text input, and a box for Text output."
interface = gr.Interface(
    fn=generate_story,
    inputs="text",
    outputs="text",
    title="My AI Storyteller ðŸ¤–",
    description="Type a sentence, and I will finish the story for you."
)

# 5. Launch it!
interface.launch(debug=True)
```

```python
# 1. Close any conflicting apps first
interface.close()

# 2. Launch in "Local Mode" (No public link, just internal)
interface.launch(share=False, debug=True)
```
