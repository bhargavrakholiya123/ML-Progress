```python
import google.generativeai as genai
from google.colab import userdata

# --- SETUP (Do this once) ---
# 1. Login
my_key = userdata.get('API_KEY')
genai.configure(api_key=my_key)
model = genai.GenerativeModel('gemini-2.5-flash')

# ==========================================
# üß± PART 1: THE DATA (Change this to build your tool)
# ==========================================
# Imagine this is your "Brain". You can paste legal laws,
# medical notes, or game rules here.
knowledge_base = {
    "topic": "Gym Trainer AI",  # Just a name for our reference
    "documents": [
        "To lose weight, you need a calorie deficit. Eat 500 fewer calories than you burn.",
        "To build muscle, lift heavy weights in the 8-12 rep range and eat 1g of protein per lb of bodyweight.",
        "Rest days are important. You should sleep 8 hours a night for muscle recovery."
    ]
}

# ==========================================
# üîç PART 2: THE RETRIEVAL ENGINE (The Logic)
# ==========================================
def retrieve_relevant_info(query, database):
    """
    This function looks at the user's question and finds the best
    matching document from our knowledge base.
    """
    results = []
    # A simple "Keyword Search" for now.
    # (In production, we use Vector Embeddings here!)
    for doc in database["documents"]:
        # If a word from the query appears in the document, we keep it.
        if any(word.lower() in doc.lower() for word in query.split()):
            results.append(doc)

    # If we found nothing, return all documents (fallback)
    if not results:
        return "\n".join(database["documents"])

    return "\n".join(results)

# ==========================================
# ü§ñ PART 3: THE GENERATOR (The Prompt Builder)
# ==========================================
def ask_assistant(user_question):
    # Step A: Retrieve the "Secret" Context
    context = retrieve_relevant_info(user_question, knowledge_base)

    # Step B: Build the Prompt (The Structure)
    # This acts as the "Manager" telling the AI how to behave.
    prompt = f"""
    You are a helpful expert assistant for: {knowledge_base['topic']}.

    Here is the strict context you must use to answer:
    ---------------------
    {context}
    ---------------------

    User Question: {user_question}

    Answer (Keep it short and friendly):
    """

    # Step C: Send to Gemini
    response = model.generate_content(prompt)
    return response.text

# ==========================================
# üéÆ PART 4: THE USER INTERFACE
# ==========================================
# Let's test it!
question = input("I am your GYM trainer")
print(f"User: {question}")
print(f"AI: {ask_assistant(question)}")
```
