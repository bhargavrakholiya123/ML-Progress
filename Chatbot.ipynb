```python
from google.colab import userdata
API = userdata.get('API_KEY')
```

```python
import google.generativeai as genai
genai.configure(api_key="API")
```

```python
model = genai.GenerativeModel("gemini-2.5-flash")

response = model.generate_content("what can i do using gemini api")
print(response.text)
```

```python
# 2. Define your "Secret Database"
# (Gemini DOES NOT know this info naturally)
private_data = """
CONFIDENTIAL INTERNAL MEMO:
Project 'Blue Sky' is actually a cover for a secret underground coffee shop.
The shop is located at 123 Fake Street, Basement Level 4.
The secret password to enter is 'Marshmallow'.
"""

# 3. The User Question
user_question = "What is the password for Project Blue Sky?"

# 4. Construct the RAG Prompt
# We paste the private data into the prompt instructions
prompt = f"""
You are a helpful assistant. Answer the user's question using ONLY the context provided below.

CONTEXT:
{private_data}

USER QUESTION:
{user_question}
"""

# 5. Ask Gemini
response = model.generate_content(prompt)

print(f"ðŸ¤– AI Answer: {response.text}")
```
